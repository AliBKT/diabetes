<font face="vazir" color="#0099cc">
Librarys
</font>
</h2>
<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    Import required libraries : 
</font>
</p>

```python
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
```

<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Read data
</font>
</h2>
<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    From array file numpy read train and test data :
</font>
</p>

```python
x_train = np.load('../x_train.npy')
y_train = np.load('../y_train.npy')
x_test = np.load('../x_test.npy')
y_test = np.load('../y_test.npy')
```

<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Train artificial neural networks
</font>
</h2>
<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    Implement logistic regression algorithms by sklearn :
</font>
</p>

```python
model = MLPClassifier(hidden_layer_sizes=128,max_iter=200)
model.fit(X=x_train,y=y_train)
```

<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Test model
</font>
</h2>

```python
# Test
y_pred_train = model.predict(X=x_train)
y_pred_test = model.predict(X=x_test)
```

<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Metrics
</font>
</h2>
<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    calculate accuracy, precision and recall for train data :
</font>
</p>

```python
accuracy_train = accuracy_score(y_true=y_train, y_pred=y_pred_train)
precision_train = precision_score(y_true=y_train, y_pred=y_pred_train)
recall_train = recall_score(y_true=y_train, y_pred=y_pred_train)
accuracy_train, precision_train, recall_train
```

<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    calculate accuracy, precision and recall for test data :
</font>
</p>

```python
accuracy_test = accuracy_score(y_true=y_test, y_pred=y_pred_test)
precision_test = precision_score(y_true=y_test, y_pred=y_pred_test)
recall_test = recall_score(y_true=y_test, y_pred=y_pred_test)
accuracy_test, precision_test, recall_test
```

<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    calculate confusion_matrix for test data : <br>
</font>
</p>

```python
ConfusionMatrix = confusion_matrix(y_true=y_test,y_pred=y_pred_test)
FP = ConfusionMatrix[0][1]
TP = ConfusionMatrix[0][0]
FN = ConfusionMatrix[1][0]
TN = ConfusionMatrix[1][1]
TP, FP, FN, TN
```
