Librarys

<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    Import required libraries : 
</font>

```python
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Read data
</font>
</h2>

<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    From array file numpy read train and test data :
</font>
</p>

```python
x_train = np.load('../x_train.npy')
y_train = np.load('../y_train.npy')
x_test = np.load('../x_test.npy')
y_test = np.load('../y_test.npy')
```

<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Train naive bayes
</font>
</h2>

<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    Implement naive bayes algoritm by sklearn
</font>
</p>

```python
model = GaussianNB()
model.fit(x_train,y_train)
```

<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Test model
</font>
</h2>

```python
y_pred_train = model.predict(X=x_train)
y_pred_test = model.predict(X=x_test)
```

<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Metrics
</font>
</h2>

<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    In this part calculate accuracy, precision and recall
</font>
</p>

```python
accuracy = accuracy_score(y_true=y_test,y_pred=y_pred_test)
precision = precision_score(y_true=y_test,y_pred=y_pred_test)
recall = recall_score(y_true=y_test,y_pred=y_pred_test)
accuracy, precision, recall
```

<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    calculate confusion_matrix : <br>
</font>
<font face="vazir" size=2>
    TP : has diabete and we recognized has diabete <br>
    FP : not has diabete and we recognized has diabete <br>
    FN : has diabete and we recognized not has diabete <br>
    TN : not has diabete we recognized not has diabete <br>
</font>
</p>

```python
ConfusionMatrix = confusion_matrix(y_true=y_test,y_pred=y_pred_test)
TP = ConfusionMatrix[0][0]
FP = ConfusionMatrix[0][1]
FN = ConfusionMatrix[1][0]
TN = ConfusionMatrix[1][1]
TP, FP, FN, TN
```
